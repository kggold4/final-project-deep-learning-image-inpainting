```python
import tensorflow as tf
from data_generators import gen_frm_pic
from util_functions import util_functions

TRGT = 4
END = 8
```

```python
data = gen_frm_pic.gen_frm_pic.generate()
```

```python
# red channel data
r_data = []

# green channel data
g_data = []

# blue channel data
b_data = []
for image in data:
    for i in range(3):

        if i == 0:
            r_temp = image[i].reshape((1, 9))[0]
            util_functions.swap(r_temp, TRGT, END)
            r_data.append(r_temp)

        if i == 1:
            g_temp = image[i].reshape((1, 9))[0]
            util_functions.swap(g_temp, TRGT, END)
            g_data.append(g_temp)
            
        if i == 2:
            b_temp = image[i].reshape((1, 9))[0]
            util_functions.swap(b_temp, TRGT, END)
            b_data.append(b_temp)

data = [r_data, g_data, b_data]
```
