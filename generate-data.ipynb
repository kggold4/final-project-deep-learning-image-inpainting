```python
import copy
import numpy as np
```

## <i>Preparing the data</i>

```python
# generate 1000 random 3d matrices for creating the dataset (1000 random 3*3 pixels images)
matrices = []
n = 1000
for _ in range(n):
    matrices.append(np.random.randint(0, 255, size=(3, 3, 3)))
```

```python
# for example the first images is:
matrices[0]
```

```python
# and the last is:
matrices[n-1]
```

## <i>Creating the dataset</i>

```python
# getting a deep copy if the images matrices
data = copy.deepcopy(matrices)

# reshape each 3d matrix in the data for the dataset
for image in data:
    image.shape = (3, 9)

# each middle value in the arrays of an image change it position to the last
for image in data:
    for mat in image:
        temp = mat[4]
        mat[4] = mat[8]
        mat[8] = temp
```
